```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Завантаження натренованої модель
model = YOLO("runs/detect/train4/weights/best.pt")

results = model.predict(
    source="manhwa_page.jpg",
    conf=0.25,
    save=False
)

# Відображення результату з використанням matplotlib
for r in results:
    # отримуємо картинку з намальованими bbox
    img = r.plot()

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(12, 8))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.show()
```
